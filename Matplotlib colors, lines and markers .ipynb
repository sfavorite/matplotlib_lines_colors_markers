```python
import matplotlib
from matplotlib import lines

lines.lineStyles.keys()

colors_array = matplotlib.colors.cnames.keys()

markers_array = matplotlib.markers.MarkerStyle.markers.keys()
```

```python
for index, line in enumerate(lines.lineStyles.keys()):
    print(line)
print(lines.lineStyles.keys())
```

```python
for index, color in enumerate(colors_array):
    print(color)
```

```python
for index, marker in enumerate(markers_array):
    print(marker)
```
